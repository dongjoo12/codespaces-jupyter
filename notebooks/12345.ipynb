```python
import matplotlib.pyplot as plt

# 백열전구와 LED 전구의 전력 소비 정의
incandescent_power = 60  # 와트 단위
led_power = 10  # # 와트 단위

# 전구의 수와 연간 사용 시간 정의
num_bulbs = 1000
hours_per_year = 365 * 24  # 하루 24시간, 1년 동안

# 연간 에너지 소비 계산 (MWh 단위)
incandescent_energy = (incandescent_power * num_bulbs * hours_per_year) / 1e6  # MWh 단위
led_energy = (led_power * num_bulbs * hours_per_year) / 1e6  # MWh 단위

# 에너지 소비 비교를 위한 막대 그래프 생성
labels = ['Incandescent Bulbs', 'LED Bulbs']
energy_consumption = [incandescent_energy, led_energy]

plt.bar(labels, energy_consumption, color=['red', 'green'])
plt.ylabel('Annual Energy Consumption (MWh)')
plt.title('Comparison of Energy Consumption: Incandescent vs LED Bulbs')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 백열전구와 LED 전구의 전력 소비 정의
백열전구_전력 = 60  # 와트 단위
led_전력 = 10  # 와트 단위

# 전구의 수와 연간 사용 시간 정의
전구_수 = 1000
연간_사용_시간 = 365 * 24  # 하루 24시간, 1년 동안

# 연간 에너지 소비 계산 (MWh 단위)
백열전구_에너지 = (백열전구_전력 * 전구_수 * 연간_사용_시간) / 1e6  # MWh 단위
led_에너지 = (led_전력 * 전구_수 * 연간_사용_시간) / 1e6  # MWh 단위

# 에너지 소비 비교를 위한 막대 그래프 생성
라벨 = ['백열전구', 'LED 전구']
에너지_소비 = [백열전구_에너지, led_에너지]

plt.bar(라벨, 에너지_소비, color=['red', 'green'])
plt.ylabel('연간 에너지 소비 (MWh)')
plt.title('에너지 소비 비교: 백열전구 vs LED 전구')
plt.show()
```
